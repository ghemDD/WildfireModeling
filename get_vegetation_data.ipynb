```python
from hda import Client

c = Client(debug=True)
```

```python
year_range = range(2017, 2022)
season_1_date = '-02-15T00:00:00.000Z'

season_2_date = '-08-15T00:00:00.000Z'
```

```python
def get_season1_dates(year):
    return (f'{year}02-15T00:00:00.000Z', f'{year}-02-16T00:00:00.000Z')
```

```python
def get_season2_date(year):
    return f'{year}{season_2_date}'
```

```python
for y in year_range:
    # Vegetation Indices - Quality-flag layer
    query = {
      "datasetId": "EO:EEA:DAT:CLMS_HRVPP_VI",
      "boundingBoxValues": [
        {
          "name": "bbox",
          "bbox": [
            -1.874850637051828,
            42.63464149876606,
            0.7006978985200955,
            43.860251947384384
          ]
        }
      ],
      "dateRangeSelectValues": [
        {
          "name": "temporal_interval",
          "start": get_season1_dates(y)[0],
          "end": get_season1_dates(y)[1]
        }
      ],
      "stringChoiceValues": [
        {
          "name": "productType",
          "value": "QFLAG2"
        }
      ]
    }

    # The following line runs the query
    veg_indices_quality_flag = c.search(query)

    # The following line prints the products returned by the query
    # print(veg_indices_quality_flag)

    #The download starts. All the products found in the query are downloaded consecutively
    res = veg_indices_quality_flag.download()
    
    # Vegetation Indices - Quality-flag layer
    query = {
      "datasetId": "EO:EEA:DAT:CLMS_HRVPP_VI",
      "boundingBoxValues": [
        {
          "name": "bbox",
          "bbox": [
            -1.874850637051828,
            42.63464149876606,
            0.7006978985200955,
            43.860251947384384
          ]
        }
      ],
      "dateRangeSelectValues": [
        {
          "name": "temporal_interval",
          "start": get_season2_date(y),
          "end": get_season2_date(y)
        }
      ],
      "stringChoiceValues": [
        {
          "name": "productType",
          "value": "QFLAG2"
        }
      ]
    }

    # The following line runs the query
    veg_indices_quality_flag = c.search(query)

    # The following line prints the products returned by the query
    # print(veg_indices_quality_flag)

    #The download starts. All the products found in the query are downloaded consecutively
    res = veg_indices_quality_flag.download()
```
